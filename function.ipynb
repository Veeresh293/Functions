1. What is the difference between a function and a method in Python?
> Function

A function is defined independently, outside of any class.

It can be called directly by its name.

It does not automatically receive an object as an argument.

Example:

def add(a, b):
    return a + b

result = add(3, 5)
print(result)

 Method

A method is a function that is defined inside a class.

It is called using an object (or class).

It usually has a special first parameter (self for instance methods), which refers to the object calling it.

Example:

class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(3, 5)
print(result)

2. Explain the concept of function arguments and parameters in Python.
>Parameters

Parameters are the variables listed in a function definition.

They act as placeholders for the values that will be passed to the function.

Example:

Here, name and message are parameters.

Arguments are the actual values passed to the function when it is called.

6. Explain the concept of generators in Python and how they are defined.
>In Python, generators are a special type of function (or expression) used to produce values one at a time, instead of creating and storing the entire sequence in memory at once.

They are mainly used for memory efficiency and lazy evaluation.

üîπ What is a Generator?

A generator:

Returns values one at a time

Uses yield instead of return

Remembers its state between calls

Is both iterable and an iterator

üîπ How Generators Are Defined
1Ô∏è‚É£ Generator Function (using yield)

A function becomes a generator when it contains the yield keyword.

def count_up(n):
    for i in range(1, n + 1):
        yield i


Using the generator:

gen = count_up(3)

print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3


Or with a loop:

for num in count_up(3):
    print(num)

2Ô∏è‚É£ Generator Expression

Similar to list comprehension, but uses parentheses instead of square brackets.

gen = (x * x for x in range(5))

for value in gen:
    print(value)

7. What are the advantages of using generators over regular functions?
>Generators have several important advantages over regular functions, especially when working with large data or continuous streams.

üîπ Advantages of Generators
1Ô∏è‚É£ Memory Efficiency

Generators do not store the entire result in memory.

They generate values one at a time, only when needed.

# List stores all values
lst = [i for i in range(1_000_000)]

# Generator produces values one by one
gen = (i for i in range(1_000_000))


‚úÖ Ideal for large datasets

2Ô∏è‚É£ Lazy Evaluation

Values are produced on demand, not in advance.

Faster program startup.

def read_file(file):
    for line in file:
        yield line

3Ô∏è‚É£ Better Performance for Large or Infinite Data

Generators can handle infinite sequences.

Regular functions cannot.

def infinite_count():
    i = 1
    while True:
        yield i

4Ô∏è‚É£ Maintains State Automatically

Generator remembers its execution state between yield calls.

No need for manual state management.

5Ô∏è‚É£ Cleaner and Simpler Code

Complex iteration logic becomes short and readable.

# Without generator
def squares(n):
    result = []
    for i in range(n):
        result.append(i*i)
    return result

# With generator
def squares(n):
    for i in range(n):
        yield i*i

6Ô∏è‚É£ Works Well with Pipelines

Generators can be chained together efficiently.

def even(nums):
    for n in nums:
        if n % 2 == 0:
            yield n

def square(nums):
    for n in nums:
        yield n*n

9. Explain the purpose and usage of the `map()` function in Python.
>The map() function in Python is used to apply a function to each item of an iterable (like a list, tuple, etc.) and return the results as a new iterator.

üîπ Purpose of map()

Transform data without using explicit loops

Write cleaner and more functional-style code

Apply the same operation to all elements

üîπ Syntax
map(function, iterable)

üîπ Basic Example
nums = [1, 2, 3, 4]

    return x * 2

map() can work on multiple iterables (stops at the shortest one).

b = [4, 5, 6]

üîπ Real-World Use Cases
1Ô∏è‚É£ Converting Data Types

nums = list(map(int, input().split()))

```python
#Question 4
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

```python
#Question 5
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # total terms to generate
        self.count = 0          # counter for generated terms
        self.a = 0              # first term
        self.b = 1              # second term

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
fib = Fibonacci(10)  # first 10 Fibonacci numbers

for num in fib:
    print(num, end=" ")
```

```python
#Question 6
def powers_of_two(n):
    """
    Generates powers of 2 from 2^0 up to 2^n
    """
    for i in range(n + 1):
        yield 2 ** i
for value in powers_of_two(5):
    print(value, end=" ")
```

```python
#Question 7
# Create a dummy file for demonstration purposes
with open("example.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")

def read_file_line_by_line(file_path):
    """
    Generator function that yields lines from a file one by one
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove newline character
file_path = "example.txt"

for line in read_file_line_by_line(file_path):
    print(line)
```
